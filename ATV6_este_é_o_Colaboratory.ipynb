```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

aparelhos = np.array([2, 4, 6, 8]).reshape(-1, 1)
custos = np.array([50, 80, 110, 140])

modelo = LinearRegression()
modelo.fit(aparelhos, custos)

previsoes = modelo.predict(aparelhos)

rmse = np.sqrt(mean_squared_error(custos, previsoes))
print(f"RMSE: {rmse}")

plt.scatter(aparelhos, custos, color='blue', label='Dados reais')
plt.plot(aparelhos, previsoes, color='red', label='Linha de regressão')
plt.xlabel('Número de Aparelhos')
plt.ylabel('Custo de Energia (R$)')
plt.title('Regressão Linear - Custo de Energia')
plt.legend()
plt.show()

```
