```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

horas = np.array([1, 5, 2, 10]).reshape(-1, 1)
risco = np.array([1, 0, 1, 0])

modelo = LogisticRegression()
modelo.fit(horas, risco)

previsoes = modelo.predict(horas)

acuracia = accuracy_score(risco, previsoes)
print(f"Acurácia: {acuracia}")

plt.scatter(horas, risco, color='blue', label='Dados reais')
plt.plot(horas, modelo.predict_proba(horas)[:, 1], color='red', label='Probabilidade de risco')
plt.xlabel('Horas de Atividade Física')
plt.ylabel('Risco de Doença Cardíaca')
plt.title('Regressão Logística - Risco de Doença Cardíaca')
plt.legend()
plt.show()
```
